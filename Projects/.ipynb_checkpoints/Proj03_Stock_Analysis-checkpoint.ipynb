### Set Up Environment

```python
#Let's go ahead and start with some imports
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# For Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# For reading stock data from yahoo
from pandas.io.data import DataReader

# For time stamps
from datetime import datetime

# For division
from __future__ import division
```

### Set Up A Data Frame for Apple Stock

```python
# Set up End and Start times for data grab
end = datetime.now()
start = datetime(end.year - 1,end.month,end.day)

df_AAPL = DataReader('AAPL','yahoo',start,end)
df_AAPL.head()
```

```python
# Let's see a historical view of the closing price
df_AAPL['Adj Close'].plot(legend=True,figsize=(10,4))
```

### Set Up A Data Frame for A Stock List

```python
# Grab all the closing prices for the tech stock list into one DataFrame

stock = ['AAPL','GOOG','MSFT','AMZN']
item = 'Adj Close'
end = datetime.now()
start = datetime(end.year - 1,end.month,end.day)

closing_df = DataReader(stock,'yahoo',start,end)[item]

closing_df.head()
```

```python
# Let's see a historical view of the closing price
closing_df.plot(legend=True,figsize=(10,4))
```
