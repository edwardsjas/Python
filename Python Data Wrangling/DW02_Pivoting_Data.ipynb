```python
# Set up the environment
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
```

```python
# Read in our sales funnel data into our DataFrame
# Use read_excel from Pandas
# Read first 10 rows
df01 = pd.read_excel('sales_funnel.xlsx','Data')
df01.head(10)
```

```python
# The simplest pivot table must have a dataframe and an index.
# In this case, let’s use the Name as our index.
pd.pivot_table(df01,index=['Name'])
```

```python
# You can have multiple indexes as well.
# In fact, most of the pivot_table args can take multiple values via a list.
pd.pivot_table(df01,index=['Name','Rep','Manager'])
```

```python
# Look at pivot by manager and rep by changing the index
# Remove Name
pd.pivot_table(df01,index=['Manager','Rep'])
```

```python
# Look at pivot by manager and rep by changing the index
# Remove Name
pd.pivot_table(df01,index=['Product','Name'])
```
